```python
#The Fjrst We Import Library to Make a Model

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
df = pd.read_csv('/Users/hendrichscullen/Desktop/Predict_Breast_Cancer/data.csv')
```

```python
# Now we can see data have a column unnecessary, it mean we must remove it
# Some causion to ensure this code infuently
# Axis == 1, mean it will drop N/A Data by Vertical Side, Axis == 0 by Horizontal
df = df.dropna(axis=1)
```

```python
x = df.drop('diagnosis' , axis=1)
```

```python
x.drop('id', axis= 1)
```

```python
y = df['diagnosis']
y
```

```python
# Split data to train and test

X_train, X_test , Y_Train , Y_Test = train_test_split(x, y , test_size= 0.33 , random_state= 2020 , stratify= y)
```

```python
# Scale Data
# You need read more and more document about benefit of Scale Data
# Before modeling, we need to “center” and “standardize” our data by scaling. We scale to control for the fact that different variables are measured on different scales. 
# We scale so that each predictor can have a “fair fight” against each other in deciding importance. See this article. 
# We also convert “y_train” from a Pandas “Series” object into a NumPy array for the model to accept the target training data later on.



import numpy as np
from sklearn.preprocessing import StandardScaler


stand = StandardScaler()
X_train_scaled = stand.fit_transform(X_train)
X_test_scaled = stand.transform(X_test)
Y_Train = np.array(Y_Train)
```

```python
# This Cell we will building model Random Forest with original data without PCA and Optimize Hyper-parameter
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score

rfc = RandomForestClassifier()
rfc.fit(X_train_scaled, Y_Train)
display(rfc.score(X_train_scaled, Y_Train))
```

```python
rfc.fit(X_test_scaled , Y_Test)
display(rfc.score(X_test_scaled, Y_Test))
```

```python
# The score above was display extract result which was trained by Model Random forest. if you need more information you can use confusion matrix to get information it

from sklearn.metrics import confusion_matrix , accuracy_score

more_inf = rfc.predict(X_train_scaled)
print("Confusion Matrix on train data: ", confusion_matrix(more_inf , Y_Train , labels=['M','B']))
print("Accuracy on train data: ", accuracy_score(more_inf , Y_Train))
```

```python
from sklearn.metrics import confusion_matrix , accuracy_score

more_inf = rfc.predict(X_test_scaled)
print("Confusion Matrix on test data: ", confusion_matrix(more_inf , Y_Test , labels=['M','B']))
print("Accuracy on test data: ", accuracy_score(more_inf , Y_Test))
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# If you want to know what top level importance of variable on This Dataset you can show it on chart
# And Check Cotribution of variable of Data

feats = {}
for feature , importance in zip(df.columns, rfc.feature_importances_):
    feats[feature] = importance

indicate = np.argsort(importance)[::-1]

importances = pd.DataFrame.from_dict(feats, orient='index').rename(columns={0: 'Gini-Importance'})
importances = importances.sort_values(by='Gini-Importance', ascending=False)
importances = importances.reset_index()
importances = importances.rename(columns={'index': 'Features'})
sns.set(font_scale = 5)
sns.set(style="whitegrid", color_codes=True, font_scale = 1.7)
fig, ax = plt.subplots()
fig.set_size_inches(30,15)
sns.barplot(x=importances['Gini-Importance'], y=importances['Features'], data=importances, color='skyblue')
plt.xlabel('Importance', fontsize=25, weight = 'bold')
plt.ylabel('Features', fontsize=25, weight = 'bold')
plt.title('Feature Importance', fontsize=25, weight = 'bold')
display(plt.show())
display(importances)
```

```python
# Now we start with PCA ( Principal Component Analysis)
from sklearn.decomposition import PCA


# PCA With 30 component
pca_test = PCA(n_components= 30)
pca_test.fit(X_train_scaled)

sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=10, ymin=0, ymax=1)
display(plt.show())




evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr
display(pca_df.head(10))
```

```python
pca = PCA(n_components= 10)
pca_test.fit(X_train_scaled)

sns.set(style='whitegrid')
plt.plot(np.cumsum(pca_test.explained_variance_ratio_))
plt.xlabel('number of components')
plt.ylabel('cumulative explained variance')
plt.axvline(linewidth=4, color='r', linestyle = '--', x=10, ymin=0, ymax=1)
display(plt.show())




evr = pca_test.explained_variance_ratio_
cvr = np.cumsum(pca_test.explained_variance_ratio_)
pca_df = pd.DataFrame()
pca_df['Cumulative Variance Ratio'] = cvr
pca_df['Explained Variance Ratio'] = evr
display(pca_df.head(10))
```

```python
pca.fit(X_train_scaled)
X_train_scaled_PCA = pca.transform(X_train_scaled)
X_test_scaled_PCA = pca.transform(X_test_scaled)
```

```python
rfc = RandomForestClassifier()
rfc.fit(X_train_scaled_PCA, Y_Train)
display(rfc.score(X_train_scaled_PCA, Y_Train))
```

```python
# Now we check this result on test data

rfc.fit(X_test_scaled_PCA, Y_Test)
display(rfc.score(X_test_scaled_PCA, Y_Test))

# You can see that it enhances your result about time excuting and percent extractly
```

```python
more_inf = rfc.predict(X_test_scaled_PCA)
print("Confusion Matrix on test data: ", confusion_matrix(more_inf , Y_Test , labels=['M','B']))
print("Accuracy on test data: ", accuracy_score(more_inf , Y_Test))
```
